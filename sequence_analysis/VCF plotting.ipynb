```python
import vcf
import os

hp = [] # heteroplasmic positions
sp = [] # all SNP positions
x = []
y = []
r = []

ycount = 0

flist = os.listdir('mito_vcf')
for files in flist:
    vcf_reader = vcf.Reader(open('mito_vcf/%s' % files, 'r'))
    for record in vcf_reader:
        strain_name = files.split('_')[0]
        if record.INFO['AC'][0] < 100: # only want heteroplasmic sites
            hp.append(record.POS)
            x.append(record.POS)
            y.append(ycount)
            r.append(100-record.INFO['AC'][0]) # make the radius bigger the more heteroplasmic it is
        sp.append(record.POS) # want all sites
    ycount = ycount+1
       
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20, 9))
ax = fig.add_subplot(312)
ay = fig.add_subplot(311)
az = fig.add_subplot(313)

ax.hist(hp, 2525, facecolor='red', alpha=0.75)
ay.hist(sp, 2525, facecolor='red', alpha=0.75)
ax.set_xlabel('Sequence Position, heteroplasmic SNPs')
ax.set_ylabel('count')
ax.set_ylim(0, 30)
ax.set_xlim(0, 25000)
ax.grid(True)
ay.set_xlabel('Sequence Position, all SNPs')
ay.set_ylabel('count')
ay.set_ylim(0, 30)
ay.set_xlim(0, 25000)
ay.grid(True)
az.scatter(x, y, c='red', s=r)
az.set_ylim(0, 30)
az.set_xlim(0, 25000)
az.set_xlabel('Sequence Position, radius indicates degree of heteroplasmy')
az.set_ylabel('Strain')
az.grid(True)
plt.show()
```
