###Analyzing and plotting mitochondiral heteroplasmy of *C. neoformans*  
Strains were aligned and analyzed as previously described.  
VCF files are loaded in and 2 pieces of information extracted: position and degree of heteroplasmy. record.INFO['AC'] is the allele count of the alternate allele, which works because of FreeBayes being run with --ploidy 100 and --pooled flags. I could change it to record.INFO['AF'] because it's a float of the frequency, but setting the radius to 1-AC works pretty well for visualization.

```python
import vcf, os
import numpy as np
import matplotlib.pyplot as plt

fig1 = plt.figure(figsize=(18, 3))
snp_plot = fig1.add_subplot(111)

snp = []

for files in flist:
    vcf_reader = vcf.Reader(open('mito_vcf/%s' % files, 'r'))
    for record in vcf_reader:
        snp.append(record.POS)

snp_plot.hist(snp, 253, facecolor='red', alpha=0.5, edgecolor='none') #this gives ~100bp per bin
snp_plot.set_xlabel('Sequence Position, all SNPs. Median read depth = %s, range = %s to %s' % (np.median(snp), min(snp), max(snp)))
snp_plot.set_ylabel('count')
snp_plot.set_ylim(0, 40)
snp_plot.set_xlim(0, 25000)
snp_plot.grid(True)
```

```python
h_snp = []

fig2 = plt.figure(figsize=(18, 3))
hsnp_plot = fig2.add_subplot(111)

for files in flist:
    vcf_reader = vcf.Reader(open('mito_vcf/%s' % files, 'r'))
    for record in vcf_reader:
        if record.INFO['AC'][0] < 100: # only want heteroplasmic sites
            h_snp.append(record.POS)

hsnp_plot.hist(snp, 253, facecolor='red', alpha=0.5, edgecolor='none') #this gives ~100bp per bin
hsnp_plot.hist(h_snp, 253, facecolor='blue', alpha=0.5, edgecolor='none')
hsnp_plot.set_xlabel('Sequence Position, heteroplasmic SNPs. Median read depth = %s, range = %s to %s' % (np.median(snp), min(snp), max(snp)))
hsnp_plot.set_ylabel('count')
hsnp_plot.set_ylim(0, 40)
hsnp_plot.set_xlim(0, 25000)
hsnp_plot.grid(True)
plt.show()
```

```python
import math
fig3 = plt.figure(figsize=(18, 5))
het_plot = fig3.add_subplot(111)

hx = []
hy = []
hr = []

ycount = 1
strain_dict = {}

flist = os.listdir('mito_vcf')
for files in flist:
    vcf_reader = vcf.Reader(open('mito_vcf/%s' % files, 'r'))
    record_name = files[:-19]
    strain_dict[ycount]=record_name
    for record in vcf_reader:
        if record.INFO['AC'][0] < 100: # only want heteroplasmic sites
            hx.append(record.POS)
            hy.append(ycount)
            radii = 2*(50-(math.fabs(record.INFO['AC'][0]-50)))
            hr.append(radii) # make the radius bigger the lower the allele count
            cov.append(record.INFO['DP'])
    ycount = ycount+1

het_plot.scatter(hx, hy, c='red', s=hr, alpha=0.5, edgecolor='none')
het_plot.set_ylim(0, 27)
het_plot.set_xlim(0, 25000)
het_plot.set_xlabel('Sequence Position, radius indicates percentage heteroplasmy')
het_plot.set_ylabel('Strain')
het_plot.grid(True)
plt.show()
```

```python
f = open('H99_mitochondria.gtf', 'r')

ex = []
ey = []

for line in f: # make a histogram that equals the exon boundries
    if line.split()[2] == 'exon':
        count =  int(line.split()[3])
        stop = int(line.split()[4])
        while count < stop:
            ex.append(count)
            ey.append(1)
            count = count+1
            
fig4 = plt.figure(figsize=(18, 1))
exon = fig4.add_subplot(111)

exon.hist(ex, 253, facecolor='green', alpha=0.7, edgecolor='none')
exon.set_xlabel('Exons')
exon.set_ylim(0, 1)
exon.set_xlim(0, 25000)
plt.show()
```
