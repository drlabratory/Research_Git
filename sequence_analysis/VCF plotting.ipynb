###Analyzing and plotting mitochondiral heteroplasmy of *C. neoformans*  
Strains were aligned and analyzed as previously described.  
VCF files are loaded in and 2 pieces of information extracted: position and degree of heteroplasmy. record.INFO['AC'] is the allele count of the alternate allele, which works because of FreeBayes being run with --ploidy 100 and --pooled flags. I could change it to record.INFO['AF'] because it's a float of the frequency, but setting the radius to 1-AC works pretty well for visualization.

```python
import vcf, os
import numpy as np

hp = [] # heteroplasmic positions
hp_d = [] # heteroplasmic read depth
sp = [] # all SNP positions
sp_d = [] # all SNP read depth
x = []
y = []
r = []
cov = []

ycount = 0

flist = os.listdir('mito_vcf')
for files in flist:
    vcf_reader = vcf.Reader(open('mito_vcf/%s' % files, 'r'))
    for record in vcf_reader:
        strain_name = files.split('_')[0]
        if record.INFO['AC'][0] < 100: # only want heteroplasmic sites
            hp.append(record.POS)
            hp_d.append(record.INFO['DP'])
            x.append(record.POS)
            y.append(ycount)
            r.append(100-record.INFO['AC'][0]) # make the radius bigger the more heteroplasmic it is
            cov.append(record.INFO['DP'])
        sp.append(record.POS) # want all sites
        sp_d.append(record.INFO['DP'])
    ycount = ycount+1

hp_dm = np.median(hp_d)
sp_dm = np.median(sp_d)

f = open('H99_mitochondria.gtf', 'r')

ex = []
ey = []

for line in f:
    if line.split()[2] == 'exon':
        count =  int(line.split()[3])
        stop = int(line.split()[4])
        while count < stop:
            ex.append(count)
            ey.append(1)
            count = count+1
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20, 12))
ax = fig.add_subplot(412)
ay = fig.add_subplot(411)
az = fig.add_subplot(413)
exon = fig.add_subplot(414)

ax.hist(hp, 2525, facecolor='red', alpha=0.75)
ay.hist(sp, 2525, facecolor='red', alpha=0.75)
ax.set_xlabel('Sequence Position, heteroplasmic SNPs. Median read depth = %s, range = %s to %s' % (hp_dm, min(hp_d), max(hp_d)))
ax.set_ylabel('count')
ax.set_ylim(0, 30)
ax.set_xlim(0, 25000)
ax.grid(True)
ay.set_xlabel('Sequence Position, all SNPs. Median read depth = %s, range = %s to %s' % (sp_dm, min(sp_d), max(sp_d)))
ay.set_ylabel('count')
ay.set_ylim(0, 30)
ay.set_xlim(0, 25000)
ay.grid(True)
az.scatter(x, y, c='red', edgecolor='red', s=r)
az.set_ylim(0, 30)
az.set_xlim(0, 25000)
az.set_xlabel('Sequence Position, radius indicates degree of heteroplasmy')
az.set_ylabel('Strain')
az.grid(True)
#az2 = az.twinx()  # too messy
#az.bar(x, y, 50, color='blue', edgecolor='none', alpha=0.2)
exon.hist(ex, 2525, facecolor='green', alpha=0.75)
exon.set_xlabel('Exons')
exon.set_ylim(0, 30)
exon.set_xlim(0, 25000)
ax.grid(True)
plt.show()
```
