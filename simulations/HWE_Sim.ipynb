#Simulating a Hardy Weinberg Equilibrium population.

This simulates population dynamics by establishing an initial population with certain allele frequencies defined by the user, randomly assigning those alleles in the population, and then mating that population. Males and females are completely arbitrary, and are not tracked from one generation to the next.

Dominant allele = 1, recessive = 0, which makes it easy to do mating outcomes via simple addition

Because the simulation *actually* mates every individual in the population, how long it takes to run scales as PopSize*Generations(*Selection) In otherwords, horribly.

It is overly reliant on "pop"ing things out of lists, and probabably naively written, but it works!

```python
from __future__ import division
import random
import matplotlib.pyplot as plt
import matplotlib 

# Initial population size
# Anything below ~2000 will regularly show significant drift
PopSize = 100

# Initial dominant allele frequency
Frequency = 0.5

# Number of Generations to simulate
Generations = 50

# Set Selection (either True or False). Only applied to homozygous recessive.
# Relatively easy to incorporate other selection scenarios. Maybe.
Selection = False

# If Selection is active, this represents the selection coefficient, and in this
# model equals the chance of survival. 0 = homozygous lethal, 1 = no selection.
SelNum = 0.9

def decision(probability):
    return random.random() > probability

def PopGen(size,freq):
    # Creating our initial population from frequency
    Alleles = []
    AlleleNum = size*2
    # Fill our list with dominant alleles (1) dependent on freq
    while len(Alleles) < AlleleNum*freq:
        Alleles.append(1)
    # Fill up the rest with recessive alleles (0)
    while len(Alleles) < AlleleNum:
        Alleles.append(0)
    #Now randomly shuffle alleles and create population
    random.shuffle(Alleles)
    population = []
    while len(Alleles)>0:
        Allele1=Alleles.pop()
        Allele2=Alleles.pop()
        eve=Allele1+Allele2
        population.append(eve)
    return population

def mate(population,sel):
    newpop=[]
    if sel:
        # This treats the selection parameter as probability of survival,
        # because in this model survival = breeding
        # I suspect this is a very wrong way to do this, python-wise
        # It also shrinks the population, creating the 
        # the potential for even greater drift effects
        for person in population:
            if person == 0:
                if decision(SelNum):
                    population.remove(person)
                    
    random.shuffle(population)
    # this assumes a self-sustaining population
    # ie - every 2 people has 2 kids
    while len(population)>1:
        male=population.pop()
        female=population.pop()
        # aa x aa
        if male+female==0:
            newpop.append(0)
            newpop.append(0)
        # Aa x aa
        elif male+female==1:
            x=0
            while x<2:
                if decision(0.5):
                    newpop.append(0)
                else:
                    newpop.append(1)
                x+=1
        # Aa x Aa
        elif male+female==2:
            kids=[0,1,1,2]
            random.shuffle(kids)
            newpop.append(kids.pop())
            newpop.append(kids.pop())
        # AA x Aa
        elif male+female==3:
            x=0
            while x<2:
                if decision(0.5):
                    newpop.append(1)
                else:
                    newpop.append(2)
                x+=1
        # AA x AA
        else:
            newpop.append(2)
            newpop.append(2)
    return newpop
    

pop=PopGen(PopSize,Frequency)

igen = 0
gen = []
homoR = []
homoD = []
het = []

while (igen < Generations):
    new=mate(pop,Selection)
    gen.append(igen)
    homoR.append(new.count(0)/len(new))
    het.append(new.count(1)/len(new))
    homoD.append(new.count(2)/len(new))
    igen = igen + 1
    pop=new

matplotlib.rcParams.update({'font.size': 22})
plt.figure(figsize=(16,8))
plt.plot(gen,homoD,marker='o', linestyle='-', color='blue', label='AA')
plt.plot(gen,het,marker='o', linestyle='-', color='green', label='Aa')
plt.plot(gen,homoR,marker='o', linestyle='-', color='yellow', label='aa')
plt.ylim(0,1)
plt.ylabel('frequency')
plt.title('Hardy-Weinberg Simulation')
plt.legend()
plt.show()
```
