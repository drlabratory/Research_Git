#Simulating a population in Hardy-Weinberg equilibrium

Thanks to [Graham Coop][1] for giving me the idea to use this as a teaching demo, and many thanks to [Joshua Shapiro][2] for setting me straight on the model and (kindly) pointing out where I was being a bonehead.

This simulates population dynamics by establishing an initial population with certain allele frequencies defined by the user, randomly assigning those alleles in the population, and then randomly mating that population. Males and females are completely arbitrary, and are not tracked from one generation to the next.

Dominant allele = 1, recessive = 0, which makes it easy to do mating outcomes via simple addition (er... except distinguishing AA x aa from Aa x Aa!, see previous comment re: bonehead)

Because this individually simulates *every* mating occurrence in the population, the time it takes to run scales as PopSize*Generations(*Selection). In other words, horribly.

```python
from __future__ import division
import random
import matplotlib.pyplot as plt
import matplotlib 

# Initial population size
# Anything below ~2000 will regularly show significant drift
PopSize = 10000

# Initial dominant allele frequency
Frequency = 0.5

# Number of Generations to simulate
Generations = 50

# Set Selection (either True or False). Only applied to homozygous recessive.
# Relatively easy to incorporate other selection scenarios. Maybe.
Selection = False

# If Selection is active, this represents the selection coefficient, and in this
# model equals the chance of survival. 0 = homozygous lethal, 1 = no selection.
SelNum = 0

def decision(probability):
    return random.random() > probability

def PopGen(size,freq):
    # Creating our initial population from frequency
    Alleles = []
    AlleleNum = size*2
    # Fill our list with dominant alleles (1) dependent on freq
    while len(Alleles) < AlleleNum*freq:
        Alleles.append(1)
    # Fill up the rest with recessive alleles (0)
    while len(Alleles) < AlleleNum:
        Alleles.append(0)
    #Now randomly shuffle alleles and create population
    random.shuffle(Alleles)
    population = []
    while len(Alleles)>0:
        Allele1=Alleles.pop()
        Allele2=Alleles.pop()
        eve=Allele1+Allele2
        population.append(eve)
    return population

def Mate(population,size,popsize,sel):
    newpop=[]
    if sel:
        # This treats the selection parameter as probability of survival,
        # I suspect this is a very wrong way to do this, python-wise
        for person in population:
            if person == 0:
                if decision(SelNum):
                    population.remove(person)
        # redefine size to avoid "out of range" errors
        size=len(population)
                    
    random.shuffle(population)
    # this now continues random mating until population size 
    # matches the initial size
    while len(newpop)<popsize:
        male=population[int(random.random()*size)]
        female=population[int(random.random()*size)]
        # aa x aa
        if male+female==0:
            newpop.append(0)
        # Aa x aa
        elif male+female==1:
            if decision(0.5):
                newpop.append(0)
            else:
                newpop.append(1)
        elif male+female==2:
            # Aa x Aa
            if male*female>0:
                kids=[0,1,1,2]
                random.shuffle(kids)
                newpop.append(kids[0])
            # AA x aa
            else:
                newpop.append(1)
        # AA x Aa
        elif male+female==3:
            if decision(0.5):
                newpop.append(1)
            else:
                newpop.append(2)
        # AA x AA
        else:
            newpop.append(2)
    return newpop
    

pop=PopGen(PopSize,Frequency)

#generation counter
igen = 0
# x axis list
gen = []
# y variable lists
homoR = []
homoD = []
het = []

while (igen < Generations):
    new=Mate(pop,PopSize,PopSize,Selection)
    gen.append(igen)
    homoR.append(new.count(0)/len(new))
    het.append(new.count(1)/len(new))
    homoD.append(new.count(2)/len(new))
    igen = igen + 1
    pop=new

matplotlib.rcParams.update({'font.size': 22})
plt.figure(figsize=(16,8))
plt.plot(gen,homoD,marker='o', linestyle='-', color='blue', label='AA')
plt.plot(gen,het,marker='o', linestyle='-', color='green', label='Aa')
plt.plot(gen,homoR,marker='o', linestyle='-', color='yellow', label='aa')
plt.ylim(0,1)
plt.ylabel('frequency')
plt.title('Hardy-Weinberg Simulation')
plt.legend()
plt.show()
```
